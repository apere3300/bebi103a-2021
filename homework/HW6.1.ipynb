```python
import numpy as np 
import scipy.stats
import scipy.linalg as spla
import bokeh.io
import pandas as pd
import iqplot
bokeh.io.output_notebook()
```

```python
def multivariate_normal_sampler(mean, covariance, n_samples=1):
    L = spla.cholesky(covariance)
    Z = np.random.normal(size=(n_samples, covariance.shape[0]))
    return Z.dot(L) + mean
```

```python
x = [6, -7, -5]
y = [-7, 13,11]
z = [-5, 11, 10]
mean = np.array([10, 15, 25]).T
covariance = np.array([x, y, z])
X = multivariate_normal_sampler(mean, covariance, n_samples=2000)
```

```python
# Set parameters
mu = 10
sigma = np.sqrt(6)
rg = np.random.default_rng()
# Draw 2000 random samples
x = rg.normal(mu, sigma, size=2000)

# Plot the histogram
p = iqplot.histogram(x, rug=False, density=True, y_axis_label="approximate PDF",)

bokeh.io.show(p)
```

```python
Tran_X = np.transpose(X) #Need to transpose since we want to make three rows x1, x2, x3
Tran_X
```

```python
X1 = Tran_X[0] #We index 0 to isolate all the values in the first row which are all our x1 values. 
X1
```

```python
# Compute theoretical CDF
x_theor = np.linspace(1, 18, 2000)
y_theor = scipy.stats.norm.cdf(x_theor, mu, sigma)

# Plot the ECDF of the 2000 x1 randomly generated numbers
p = iqplot.ecdf(X1, marker_kwargs={"fill_color": None},)

p.line(
    x=x_theor, y=y_theor, line_width=2, line_color="orange",
)

bokeh.io.show(p)
```
