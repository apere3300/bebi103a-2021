4. Make a plot of the signal from the electrophysiological recording versus time for four of the channels during one second of recording.

* Needed: Signal vs time, during 1 second of time, 4 channels (so 4 plots)

```python
import numpy as np
# experiment 1 (0 in Python)
experiment = file.experiments[0]
# recording 1 
recording = experiment.recordings[0]

analog_signals = recording.analog_signals
events_data = recording.events
spiketrains = recording.spiketrains

# tracking_data are accessible only using binary format
tracking_data = recording.tracking

# plot analog signal of channel 1 to 4. Since we start with 0 we need to use singal [0] to plot channel 1 

signals = analog_signals[0]
x = range(60)
fig_an, ax_an = plt.subplots()
ax_an.plot(signals.times, signals.signal[0])
plt.xlabel('time (s)')
plt.ylabel('signal')
ax_an.set_xlim([0, 1])

signals = analog_signals[0]
x = range(60)
fig_an, ax_an = plt.subplots()
ax_an.plot(signals.times, signals.signal[1])
plt.xlabel('time (s)')
plt.ylabel('signal')
ax_an.set_xlim([0, 1])

signals = analog_signals[0]
fig_an, ax_an = plt.subplots()
ax_an.plot(signals.times, signals.signal[2])
plt.xlabel('time (s)')
plt.ylabel('signal')
ax_an.set_xlim([0, 1])

signals = analog_signals[0]
fig_an, ax_an = plt.subplots()
ax_an.plot(signals.times, signals.signal[3])
plt.xlabel('time (s)')
plt.ylabel('signal')
ax_an.set_xlim([0, 1])

signals = analog_signals[0]
x = range(60)
fig_an, ax_an = plt.subplots()
ax_an.plot(signals.times, signals.signal[34])
plt.xlabel('time (s)')
plt.ylabel('signal')
ax_an.set_xlim([0, 1])

signals = analog_signals[0]
x = range(60)
fig_an, ax_an = plt.subplots()
ax_an.plot(signals.times, signals.signal[38])
plt.xlabel('time (s)')
plt.ylabel('signal')
ax_an.set_xlim([0, 1])


plt.show()
```
